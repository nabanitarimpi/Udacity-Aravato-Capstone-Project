Next we replace all the entries of the `azdias` dataframe representing unknown values by nan.

```python
def replace_unknwn(attr_df, df):
    
    """
    This function replaces unknown values in the dataframe 'df' by nan.
    
    Parameters
    -------------
    attr_df : pandas dataframe
         the dataframe containing information about all the features of the dataframe 'df'
         
    df : pandas dataframe
         the input dataframe containing unknown values
         
     
    Returns
    ----------
    df : pandas dataframe
         the dataframe with all unknown values replaced by nan 
    
    """
    
    start_time = time()
    for attr in attr_df['Attribute']:
        if attr in df.columns:
            print("now scanning through column {}".format(attr)) 
            rows = df[attr].isin(attr_df.loc[attr_df["Attribute"]==attr, 'Value'].iloc[0])
            df.at[rows, attr] = np.nan
            
    elapsed_time = time() - start_time     
    print()
    print('total elapsed time : {}'.format(elapsed_time))
    return df
```

### (F) Outlier detection

Finally, we examine the number of outliers in each row and column of our dataframe separately. If we find any row/column with large number of outliers, we will drop it.

To determine outliers, we first calculate z_score of each entry in the dataframe.

```python
z_score = np.abs(stats.zscore(azdias))
```

We mark entries with z_score > 3 as outliers and extract their position in the dataframe.

```python
outlier_pos_row, outlier_pos_col = np.where(z_score > 3.0)
```

The first and second entries denote row and column numbers respectively.

```python
outlier_pos_row 
```

```python
outlier_pos_col
```

Next we compute how many times a particular row/column appears in the above two arrays. This is because the more number of times it appears in these arrays, the more outliers it has.

```python
# returns count for each unique row element 
row, row_count = np.unique(outlier_pos_row, return_counts=True)

# returns count for each unique column element 
col, col_count = np.unique(outlier_pos_col, return_counts=True)
```

We construct a dictionary whose keys are unique rows/columns and values are the corresponding percentage of outliers in that row/column.

```python
# for rows
outlier_row_count = {k:(v/azdias.shape[1] * 100) for k,v in zip(row, row_count)}

# for columns
outlier_col_count = {k:(v/azdias.shape[0] * 100) for k,v in zip(col, col_count)}
```

```python
outlier_row_count = {k:v for k,v in sorted(outlier_row_count.items(), key=lambda item:item[1], reverse=True)}
outlier_col_count = {k:v for k,v in sorted(outlier_col_count.items(), key=lambda item:item[1], reverse=True)}
```

### (G) Combining all data cleaning steps into a single function

In this step, we will combine all our data cleaning steps into a single function. Since both of our dataframes `azdias` and `customers` have exactly the same features, we can directly apply all these steps on the `customers` dataframe also.

```python
def cleaning_func(cols_to_drop, df, feat_df=feature_val_unknwn):
    
    """
    The function to clean an input dataframe using the following steps :
    1. Drop columns with large fraction(>20%) of null values, high correlation coefficients(>80%) and large number
       of unique categories(>15).
    2. Drop rows with large fraction(>20%) of null values. 
    3. Replace unknown values in the dataframe by nan.
    4. Feature engineering of the categorical columns.
    5. Drop duplicate rows if present.
    
    Parameters
    -------------
    cols_to_drop : numpy array
        the columns to be dropped from the input dataframe
    
    df : pandas dataframe
        the input dataframe
        
    feat_df : pandas dataframe
        the dataframe containg descriptions of features present in the dataset 'df'
        
        
    Returns
    ----------
    df : pandas dataframe
        the cleaned dataframe
    
    
    """
    
    print("Initial shape of the dataframe : {}".format(df.shape))
    
    # drop columns with large number of missing values and also which are highly correlated
    df.drop(cols_to_drop, axis=1, inplace=True)
    print("The shape of the dataframe after dropping columns : {}".format(df.shape))
    
    # drop rows with large number of missing values
    df_null_rows = df.isnull().sum(axis=1).sort_values(ascending=False)/df.shape[1] * 100
    rows_to_drop = df_null_rows[df_null_rows > 20.].index
    df.drop(rows_to_drop, axis=0, inplace=True)
    print("The shape of the dataframe after dropping rows : {}".format(df.shape))
    
    # replacing all the unknown values with nan
    start_time = time()
    for attr in feat_df['Attribute']:
        if attr in df.columns:
            print("Now scanning through column {}".format(attr)) 
            rows = df[attr].isin(feat_df.loc[feat_df["Attribute"]==attr, 'Value'].iloc[0])
            df.at[rows, attr] = np.nan
            
    elapsed_time = time() - start_time     
    print('Total elapsed time : {}'.format(elapsed_time))
    
    # check whether we can drop more rows based on the same criteria after replacing unknowns by nan
    df_null_new_rows = df.isnull().sum(axis=1).sort_values(ascending=False)/df.shape[1] * 100
    new_rows_to_drop = df_null_new_rows[df_null_new_rows > 20.].index
    df.drop(new_rows_to_drop, axis=0, inplace=True)
    print("The shape of the dataframe after dropping rows for the second time : {}".format(df.shape))
    
    # set the index of the dataframe appropriately
    df.index = range(len(df))
    
    # columns with mixed value types
    df['CAMEO_DEUG_2015'].replace(['4', '3', '7', '2', '8', '9', '6', '5', '1', 'X'], 
                                     [4.0, 3.0, 7.0, 2.0, 8.0, 9.0, 6.0, 5.0, 1.0, np.nan], inplace=True)
    
    # encode columns with binary categories
    df['OST_WEST_KZ'] = df['OST_WEST_KZ'].replace(['O', 'W'], [0,1])
    
    # drop duplicate rows (if any)
    df.drop_duplicates(inplace=True)
    print("The shape of the dataframe after removing duplicate rows : {}".format(df.shape))

    return df
```

Therefore, `azdias` has three extra columns as compared to `customers` with null values. So we can use the list `cols_with_nans_1` to impute null values in both the dataframes.

For imputation of null values, we use the following logic:

- For each of the columns with null values, we first determine percentage of each unique value in that column (in a descending order) which can be thought of as frequency of that particular value.
- If the top most value (*i.e.*, the value with the highest percentage) has a frequency > 60., we will fill up null values in that column using the strategy 'most_frequent'.
- If the top most value has a frequency < 60., we will fill up null values in that column using the strategy 'median'.

```python
cols_for_mf, cols_for_me = [], []

for col in cols_with_nans_1:
    print("% of values for top 3 entries in {}".format(col))
    print(azdias[col].value_counts().values[:3]/len(azdias) * 100)
    print()
    
    # cols to be filled in using 'most_frequent' strategy
    if azdias[col].value_counts().values[0]/len(azdias) * 100 > 60.:
        cols_for_mf.append(col)
    # cols to be filled in using 'median' strategy    
    else:
        cols_for_me.append(col)
```

```python
len(cols_for_mf), len(cols_for_me)
```

```python
# imputer objects 
imputer_mf = Imputer(strategy='most_frequent', axis=1, copy=False)
imputer_me = Imputer(strategy='median', axis=1, copy=False)
```

```python
# imputation using 'most_frequent' strategy
azdias_imputed_df_1 = pd.DataFrame(imputer_mf.fit_transform(azdias[cols_for_mf]), columns=cols_for_mf)
customers_imputed_df_1 = pd.DataFrame(imputer_mf.transform(customers[cols_for_mf]), columns=cols_for_mf)

# imputation using 'median' strategy
azdias_imputed_df_2 = pd.DataFrame(imputer_me.fit_transform(azdias[cols_for_me]), columns=cols_for_me)
customers_imputed_df_2 = pd.DataFrame(imputer_me.transform(customers[cols_for_me]), columns=cols_for_me)
```

### (B)  Formation of clusters

In this section, we will use clustering techniques to group similar individuals in the feature space. Our aim is to form clusters using the general population data and then map it to the customer data. This will enable us to identify which part(s) of the population is more likely to become customers of the company.

Here we use k-means clustering algorithm to find clusters in the feature space. One important parameter of this algorithm is n_clusters which is the final number of clusters we want to form. One way to determine this number is to use the `elbow plot`. This is a plot of `within cluster sum of squared metric` (wcss) as a function of cluster number. As the number of cluster increases, wcss which is basically the sum of squared distances between the centroid and other datapoints in a cluster, decreases. The optimal number of clusters can be chosen from the plot as a point beyond which the wcss metric either does not change anymore or the change is not substantial.

```python
# list in which we will store wcss values
wcss = []

start_time = time()
# run a loop on n_clusters within a range
for i in range(1, 22, 2):
    # construct the KMEANS object with the corresponding number of clusters
    kmeans = KMeans(n_clusters=i)
    print("fitting with n_cluster = {}".format(i))
    # fit the dataframe azdias_pca to it
    kmeans.fit(azdias_pca)
    # store the value of wcss metric
    wcss.append(kmeans.inertia_)
    
elapsed_time = time() - start_time
print()
print("total elapsed time : {}".format(elapsed_time))
```

```python
plt.plot(range(1, 22, 2), wcss, marker="o")
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
plt.savefig('elbow_plot.png')
```

```python
def cleaning_func2(cols_to_drop, df, feat_df=feature_val_unknwn):
    
    """
    The function to clean an input dataframe using the following steps :
    1. Drop columns with large fraction(>20%) of null values, high correlation coefficients(>80%) and large number
       of unique categories(>15).
    2. Drop rows with large fraction(>20%) of null values. 
    3. Feature engineering of the categorical columns.
    4. Drop duplicate rows if present.
    
    Parameters
    -------------
    cols_to_drop : numpy array
        the columns to be dropped from the input dataframe
    
    df : pandas dataframe
        the input dataframe
        
    feat_df : pandas dataframe
        the dataframe containg descriptions of features present in the dataset 'df'
        
        
    Returns
    ----------
    df : pandas dataframe
        the cleaned dataframe
    
    
    """
    
    print("Initial shape of the dataframe : {}".format(df.shape))
    
    # drop columns with large number of missing values and also which are highly correlated
    df.drop(cols_to_drop, axis=1, inplace=True)
    print("The shape of the dataframe after dropping columns : {}".format(df.shape))
    
    # drop rows with large number of missing values
    df_null_rows = df.isnull().sum(axis=1).sort_values(ascending=False)/df.shape[1] * 100
    rows_to_drop = df_null_rows[df_null_rows > 20.].index
    df.drop(rows_to_drop, axis=0, inplace=True)
    print("The shape of the dataframe after dropping rows : {}".format(df.shape))
    
    # set the index of the dataframe appropriately
    df.index = range(len(df))
    
    # columns with mixed value types
    df['CAMEO_DEUG_2015'].replace(['4', '3', '7', '2', '8', '9', '6', '5', '1', 'X'], 
                                     [4.0, 3.0, 7.0, 2.0, 8.0, 9.0, 6.0, 5.0, 1.0, np.nan], inplace=True)
    
    # encode columns with binary categories
    df['OST_WEST_KZ'] = df['OST_WEST_KZ'].replace(['O', 'W'], [0,1])
    
    # drop duplicate rows (if any)
    df.drop_duplicates(inplace=True)
    print("The shape of the dataframe after removing duplicate rows : {}".format(df.shape))

    return df
```

### (C) Build the model

In this part, we build a model that will be able to predict how likely an individual is to respond to the campaign. We build a ML pipeline with the following steps : 

- null value imputation
- feature scaling
- classifier 

We then apply grid search technique on this pipeline object to find out which set of parameters gives the best score on the validation data. Both imputation and feature scaling are added within the pipeline to avoid data leakage during grid search. Since the dataset is highly imbalanced, accuracy is not a good measure in this case to evalute the model's performance on the validation set. We are mainly interested in predicting how likely an individual is to response to the campaigning so that pepole with high probabilities can be targeted during the campaign. Class to which an individual belongs is not important. Therefore, we will compute area under the ROC curve to evaluate model's performance. 

```python
def build_model(clf_list, param_list):
    
    """
    The function to build a classification model. This function makes use of a data pipeline containing the 
    following three steps :
    
    1. imputer : to impute null values 
    2. scaler : to scale all valuesso that they lie within the same range
    3. classifier : the classifier model 
    
    The grid search technique is then applied on this pipeline object for prespecified values of different
    parameters associated with the pipeline.
    
    Parameters
    ------------
    clf_list : list
        a list of classifier objects that will be used in the pipeline
        
    param_list : list
        a list of dictionaries with parameters names (str) as keys and lists of parameter settings to try as 
        values at the time of grid search  
        
    
    Returns
    ----------
    gridcv_list : list
         a list of GridSearchCV objects 
    
    """
    
    # list to store GridSearchCV objects
    gridcv_list = []
    
    # for each classifier object and corresponding dict of paramets
    for clf,param_grid in zip(clf_list, param_list):
        # create the pipeline object
        pipeline = Pipeline([
                     ('imputer', Imputer(axis=1, copy=False)),
                     ('scaler', StandardScaler()),
                     ('classifier', clf)
                    ])
        
        # create the GridSearchCV object 
        gridcv = GridSearchCV(pipeline, param_grid, scoring='roc_auc', cv=5)
        # append the object to the list
        gridcv_list.append(gridcv)
  
    return gridcv_list    
```

```python
def fit_model(gridcv_list, clf_list, X, y):
    
    """
    The function to train a model. It fits gridcv estimators to the training data (X,y). The function prints the 
    parameters set that produces the highest score on the left out data (i.e. the validation data) during grid 
    search and the corresponding score.
    
    Parameters
    ------------
    gridcv_list : list
        a list of gridcv estimators 
          
    clf_list : list
        a list of classifiers
        
    X : pandas dataframe
        the feature matrix
        
    y : numpy array
        the target variable
        
        
    Returns
    ----------
    None
    
    """
    
    for i, gridcv in enumerate(gridcv_list):
        print("the current model is : {}".format(clf_list[i]))
        print("fitting the model!")
        start_time = time()
        gridcv.fit(X, y)
        print("the best parameter is : {}".format(gridcv.best_params_))
        print("the best training score is : {}".format(gridcv.best_score_))
        elapsed_time = time() - start_time
        print("total time to fit : {}".format(elapsed_time))
        print()
```

```python
# create a list of classifiers that will used during grid search
clf_list = [LogisticRegression(class_weight='balanced'), RandomForestClassifier(class_weight='balanced'), 
            AdaBoostClassifier(), XGBClassifier()]

# create a corresponding list of dictionaries of parameters over which grid search will be performed
param_list = [{'imputer__strategy': ["median", 'most_frequent'],
              'classifier__C':[0.01, 0.05, 0.1]},
              {'imputer__strategy': ["median", 'most_frequent'],
              'classifier__n_estimators': [20, 50, 80, 100],
               'classifier__criterion': ['gini', 'entropy']},
              {'imputer__strategy': ["median", 'most_frequent'],
               'classifier__n_estimators': [20, 50, 80, 100],
               'classifier__learning_rate': [0.01, 0.05, 0.1]},
              {'imputer__strategy': ["median", 'most_frequent'],
               'classifier__learning_rate': [0.01, 0.05, 0.1]}]
```

```python
# create the list of gridcv estimators
gridcv_list = build_model(clf_list=clf_list, param_list=param_list)
```

### Further improving the model

Now that we have chosen our model, let's try to improve it's performance further. We have already seen that our dataset is severely imbalanced with very few samples of the minority class (*i.e.*, the class with RESPONSE=1). A simple approach to handle this situation is to put more weights on the positive class as compared to the negative class. XGBClassifier has a parameter called 'scale_pos_weight' which can be used to put more weight on the negative class to achieve a balance between both the classes. A common choice for the weight is the ratio between the majority and the minority class.

```python
# compute the weight factor
weight = np.round((y==0).sum()/(y==1).sum())

# fit the model to the training set
clf_new = XGBClassifier(learning_rate=0.01, scale_pos_weight=weight)
clf_new.fit(X_train_scaled, y_train)
```

```python
# compute probability on the test set
new_pred = clf_new.predict_proba(X_test_scaled)

# finally compute the ROC AUC on the test set
print("ROC AUC score for the test set is : {}".format(roc_auc_score(y_test, new_pred[:, 1])))
```

Adjusting the weight parameter does not improve the model much.
